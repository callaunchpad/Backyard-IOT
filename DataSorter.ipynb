```python
import json
import os
import shutil
```

```python
"""Prepare Data"""
images_path = "eccv_18_all_images_sm"
labels_path = "CaltechCameraTrapsECCV18.json"

with open(labels_path) as json_file:
    data = json.load(json_file)
    
path_lst = next(os.walk(images_path))[2]

num_images = len(path_lst)
print(num_images)

data['categories']
```

```python
data['annotations']
```

```python
idx_to_animal = {}
for category in data['categories']:
    idx_to_animal[category['id']] = category['name']

idx_to_animal
```

```python
image_to_animal = {}
for annotation in data['annotations']:
    image_to_animal[annotation['image_id']] = idx_to_animal[annotation['category_id']]
```

```python
image_to_animal['5a197af2-23d2-11e8-a6a3-ec086b02610b']
```

```python
FROM_PATH = 'eccv_18_all_images_sm/'
TO_PATH = 'data/'
def copy(image):
    category = image_to_animal[image.split('.')[0]]
    if category not in seen:
        os.mkdir(TO_PATH + category)
        seen.add(category)
    shutil.copy(FROM_PATH + image, TO_PATH + category + '/' + image)
```

```python
#already seen categories
seen = set()

for i, image in enumerate(path_lst):
    copy(image)
    
    if i % 1000 == 0:
        print('{}% done!'.format(i / num_images * 100))
```
